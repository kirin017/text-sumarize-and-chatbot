```python
# Cell 1: Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datasets import load_dataset
import warnings
warnings.filterwarnings('ignore')
```

```python
#Load dataset

dataset = load_dataset('cnn_dailymail', '3.0.0', split='train[:1000]')
print(f"Số sample: {len(dataset)}")

print(dataset.features)
```

```python
sample = dataset[0]

print(f"ID {sample['id']}")
print(f"Article length: {len(sample['article'])}")
print(f"Summary length: {len(sample['highlights'])}")

print("\n--- ARTICLE ---")
print(sample["article"])
print("\n--- SUMMARY ---")
print(sample["highlights"])
```

```python
article_word_lengths = [len(text['article'].split()) for text in dataset]
summary_word_lengths = [len(text["highlights"].split()) for text in dataset]

print(f"Article - Mean: {np.mean(article_word_lengths):.1f}, "
      f"Max: {max(article_word_lengths)}, Min: {min(article_word_lengths)}")
print(f"Summary - Mean: {np.mean(summary_word_lengths):.1f}, "
      f"Max: {max(summary_word_lengths)}, Min: {min(summary_word_lengths)}")

# Vẽ histogram
plt.figure(figsize=(12, 4))

plt.subplot(1, 2, 1)
plt.hist(article_word_lengths, bins=30, alpha=0.7, color='blue')
plt.title('Phân bố độ dài Article (từ)')
plt.xlabel('Số từ')
plt.ylabel('Tần suất')

plt.subplot(1, 2, 2)
plt.hist(summary_word_lengths, bins=30, alpha=0.7, color='red')
plt.title('Phân bố độ dài Summary (từ)')
plt.xlabel('Số từ')
plt.ylabel('Tần suất')

plt.tight_layout()
plt.show()
```
